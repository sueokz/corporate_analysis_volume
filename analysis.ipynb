```python
import sys
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import japanize_matplotlib
from datetime import datetime
```

```python
#日本株(トヨタ)
stock_code = '7203.T'
```

```python
# 株情報の取得開始日を指定する
start = '2021-04-1'

# 株情報の取得終了日を指定する
end = '2023-04-01'

# 期間を指定して株情報を取得する
df = yf.download(tickers=stock_code, start=start, end=end)

df
```

```python
#52週高値とvolumeとsma
df["52w_high"] = df["High"].rolling(window=52).max()
df["52w_volume"] = df["Volume"].rolling(window=52).max()
df["sma10_volume"] = df["Volume"].rolling(window=10).mean()
```

```python
df
```

```python
#signal1_現在値がhighの何%以内
df["price_zone"]=df["Open"]/df["52w_high"]*100
df["signal_1"] = df['price_zone'] > 25
df
```

```python
if str(df["signal_1"].iloc[-1]).lower() == "true":
    print("高値圏にいます!")
else:
    print("値位置シグナル✘")
```

```python
#signal2_現在volumeが平均出来高より上_signal2
df["signal_2"]=df["Volume"]>df["sma10_volume"]
df
```

```python
if str(df["signal_2"].iloc[-1]).lower() == "true":
    print("出来高が平均以上です!")
else:
    print("出来高シグナル✘")
```

```python
#週足の取得
df_w = yf.download(tickers=stock_code, start=start, end=end,interval='1wk')
df_w
```

```python
#signal3_週足の出来高が前週と比べて2倍
df_w['前週との差']=df_w['Volume'].pct_change( )
df_w["signal_3"]=df_w['前週との差']>1
df_w
```

```python
if str(df_w["signal_3"].iloc[-1]).lower() == "true":
    print("週足の出来高が前週と比べて2倍です!")
else:
    print("週足の出来高シグナル✘")
```
